Criando uma funÃ§Ã£o para obter os informaÃ§Ã£o do dataframe

```python
def df_informacoes(df):
    df_info = pd.DataFrame({'NÃ£o Nulos': df.notnull().sum(),
                'Nulos': df.isnull().sum(),
                '% Nulos': ((df.isnull().sum()/df.notnull().sum())*100).round(2),                         
                'Unicos': df.nunique(),                  
                'Dtype': df.dtypes})
    df_info.index.name = 'Columns'
    return df_info
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker



variables = ['Taxa_de_natalidade', 'Crescimento_do_pib', 'Pib_per_capita',
       'Crescimento_do_pib_per_capita', 'Pib', 'Taxa_de_mortalidade',
       'Despesa_governamental_em_educacao', 'Expectativa_de_vida_ao_nascer',
       'Crescimento_populacao_urbana', 'Populacao_urbana_percentual',
       'Populacao_urbana_total', 'Crescimento_populacao_rural',
       'Populacao_rural_percentual', 'Populacao_rural_total',
       'Despesa_em_pesquisa_e_desenvolvimento', 'Gasto_governamental_saude',
       'Despesa_militar_percentual', 'Densidade_populacional',
       'Populacao_total']


# Cria uma matriz de subplots (2 linhas, 5 colunas)
fig, axs = plt.subplots(nrows=5, ncols=4, figsize=(20, 12))

# Cria um Ã­ndice para iterar sobre as variÃ¡veis
var_index = 0

# Itera sobre as linhas e colunas dos subplots
for i in range(5):
    for j in range(4):
        # Se todas as variÃ¡veis tiverem sido plotadas, sai do loop
        if var_index == len(variables):
            break
        
        # Plota a variÃ¡vel atual em um subplot especÃ­fico
        axs[i,j].plot(ind['Ano'], ind[variables[var_index]])
        axs[i,j].set_title(variables[var_index])
        axs[i,j].set_xlabel('Ano')
        axs[i,j].set_ylabel(variables[var_index])

        # Define o formato dos tick labels no eixo y
        if "Populacao" in variables[var_index]:
            # Para variÃ¡veis de populaÃ§Ã£o, formata em milhÃµes
            formatter = ticker.FuncFormatter(lambda x, pos: '{:.1f} Mi'.format(x/1e6))
            axs[i,j].yaxis.set_major_formatter(formatter)
        elif "Pib" in variables[var_index] and variables[var_index] != 'Pib_per_capita':
            # Para variÃ¡veis de PIB, formata em trilhÃµes
            formatter = ticker.FuncFormatter(lambda x, pos: '{:.2f} Tri'.format(x/1e12))
            axs[i,j].yaxis.set_major_formatter(formatter)
        elif "Pib_per_capita" in variables[var_index]:
            # Para variÃ¡veis de PIB per capita
            formatter = ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x))
            axs[i,j].yaxis.set_major_formatter(formatter)    
        else:
            # Para outras variÃ¡veis, usa a notaÃ§Ã£o padrÃ£o
            formatter = ticker.ScalarFormatter()
            formatter.set_powerlimits((-3,3))
            axs[i,j].yaxis.set_major_formatter(formatter)
        
        # Incrementa o Ã­ndice da variÃ¡vel
        var_index += 1
# Ajusta os subplots e o espaÃ§amento
plt.tight_layout()

# Mostra o grÃ¡fico
plt.show()
```

```python
# Selecionar apenas as colunas desejadas
ind_despesas = ind[['Nome_do_pais', 'Ano', 'Despesa_em_pesquisa_e_desenvolvimento', 'Gasto_governamental_saude', 'Despesa_governamental_em_educacao']]

# Selecionar apenas os anos maiores ou iguais a 1995
ind_despesas = ind_despesas[ind_despesas['Ano'] >= 1995]

# Fazer a interpolaÃ§Ã£o linear nos valores faltantes
ind_despesas_interpolated = ind_despesas.interpolate()

# Adicionar asteriscos aos novos valores criados pela interpolaÃ§Ã£o
for col in ind_despesas.columns:
    if ind_despesas[col].dtype != 'object': # verificar se a coluna Ã© numÃ©rica
        new_values_mask = (ind_despesas_interpolated[col] != ind_despesas[col])
        ind_despesas.loc[new_values_mask, col] = ind_despesas_interpolated[col].round(4).astype(str) + '*'

# Remover asteriscos dos valores nulos
ind_despesas = ind_despesas.replace('nan*', np.nan)

# Exibir a nova tabela
ind_despesas.head(31)

# Exportar a nova tabela em um arquivo CSV
ind_despesas.to_csv('despesas_ind_interp.csv', index=False)
```

```python
# Selecionar apenas as colunas desejadas
ind_despesas1 = ind[['Nome_do_pais', 'Ano', 'Despesa_em_pesquisa_e_desenvolvimento', 'Gasto_governamental_saude', 'Despesa_governamental_em_educacao']]

# Selecionar apenas os anos maiores ou iguais a 1995
ind_despesas1 = ind_despesas1[ind_despesas1['Ano'] >= 1995]

# Fazer a interpolaÃ§Ã£o linear nos valores faltantes
ind_despesas_interpolated1 = ind_despesas1.interpolate()

# Converter colunas numÃ©rcias para float
cols_numeric = ['Despesa_em_pesquisa_e_desenvolvimento', 'Gasto_governamental_saude', 'Despesa_governamental_em_educacao']
ind_despesas_interpolated1[cols_numeric] = ind_despesas_interpolated1[cols_numeric].astype(float)

# Exibir a nova tabela
ind_despesas_interpolated1.head(31)
```

```python
# Criar dicionÃ¡rio com novos nomes das colunas
new_cols = {'Despesa_em_pesquisa_e_desenvolvimento': 'P&D', 
            'Gasto_governamental_saude': 'SaÃºde', 
            'Despesa_governamental_em_educacao': 'EducaÃ§Ã£o'}

# Renomear as colunas
ind_despesas_interpolated1 = ind_despesas_interpolated1.rename(columns=new_cols)

# Selecionar colunas renomeadas para o grÃ¡fico
cols = list(new_cols.values())

# Plotar grÃ¡fico de linhas
ind_despesas_interpolated1.plot(x='Ano', y=cols, kind='line', figsize=(10,6))

# Definir tÃ­tulo e legendas dos eixos
plt.title('Estimativa da Despesas da India em P&D, saÃºde e educaÃ§Ã£o (1995-2022)', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Despesa (% do PIB)', fontsize=12)

# Colocar legenda fora do grÃ¡fico
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

# Exibir grÃ¡fico
plt.show()

```
