Criando uma função para obter os informação do dataframe

```python
def df_informacoes(df):
    df_info = pd.DataFrame({'Não Nulos': df.notnull().sum(),
                'Nulos': df.isnull().sum(),
                '% Nulos': ((df.isnull().sum()/df.notnull().sum())*100).round(2),                         
                'Unicos': df.nunique(),                  
                'Dtype': df.dtypes})
    df_info.index.name = 'Columns'
    return df_info
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker



variables = ['Taxa_de_natalidade', 'Crescimento_do_pib', 'Pib_per_capita',
       'Crescimento_do_pib_per_capita', 'Pib', 'Taxa_de_mortalidade',
       'Despesa_governamental_em_educacao', 'Expectativa_de_vida_ao_nascer',
       'Crescimento_populacao_urbana', 'Populacao_urbana_percentual',
       'Populacao_urbana_total', 'Crescimento_populacao_rural',
       'Populacao_rural_percentual', 'Populacao_rural_total',
       'Despesa_em_pesquisa_e_desenvolvimento', 'Gasto_governamental_saude',
       'Despesa_militar_percentual', 'Densidade_populacional',
       'Populacao_total']


# Cria uma matriz de subplots (2 linhas, 5 colunas)
fig, axs = plt.subplots(nrows=5, ncols=4, figsize=(20, 12))

# Cria um índice para iterar sobre as variáveis
var_index = 0

# Itera sobre as linhas e colunas dos subplots
for i in range(5):
    for j in range(4):
        # Se todas as variáveis tiverem sido plotadas, sai do loop
        if var_index == len(variables):
            break
        
        # Plota a variável atual em um subplot específico
        axs[i,j].plot(chi['Ano'], chi[variables[var_index]])
        axs[i,j].set_title(variables[var_index])
        axs[i,j].set_xlabel('Ano')
        axs[i,j].set_ylabel(variables[var_index])

        # Define o formato dos tick labels no eixo y
        if "Populacao" in variables[var_index]:
            # Para variáveis de população, formata em milhões
            formatter = ticker.FuncFormatter(lambda x, pos: '{:.1f} Mi'.format(x/1e6))
            axs[i,j].yaxis.set_major_formatter(formatter)
        elif "Pib" in variables[var_index] and variables[var_index] != 'Pib_per_capita':
            # Para variáveis de PIB, formata em trilhões
            formatter = ticker.FuncFormatter(lambda x, pos: '{:.2f} Tri'.format(x/1e12))
            axs[i,j].yaxis.set_major_formatter(formatter)
        elif "Pib_per_capita" in variables[var_index]:
            # Para variáveis de PIB per capita
            formatter = ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x))
            axs[i,j].yaxis.set_major_formatter(formatter)    
        else:
            # Para outras variáveis, usa a notação padrão
            formatter = ticker.ScalarFormatter()
            formatter.set_powerlimits((-3,3))
            axs[i,j].yaxis.set_major_formatter(formatter)
        
        # Incrementa o índice da variável
        var_index += 1
# Ajusta os subplots e o espaçamento
plt.tight_layout()

# Mostra o gráfico
plt.show()
```

```python
# Selecionar apenas as colunas desejadas
chi_despesas = chi[['Nome_do_pais', 'Ano', 'Despesa_em_pesquisa_e_desenvolvimento', 'Gasto_governamental_saude', 'Despesa_governamental_em_educacao']]

# Selecionar apenas os anos maiores ou iguais a 1995
chi_despesas = chi_despesas[chi_despesas['Ano'] >= 1995]

# Fazer a interpolação linear nos valores faltantes
chi_despesas_interpolated = chi_despesas.interpolate()

# Adicionar asteriscos aos novos valores criados pela interpolação
for col in chi_despesas.columns:
    if chi_despesas[col].dtype != 'object': # verificar se a coluna é numérica
        new_values_mask = (chi_despesas_interpolated[col] != chi_despesas[col])
        chi_despesas.loc[new_values_mask, col] = chi_despesas_interpolated[col].round(4).astype(str) + '*'

# Remover asteriscos dos valores nulos
chi_despesas = chi_despesas.replace('nan*', np.nan)

# Exibir a nova tabela
chi_despesas.head(31)

# Exportar a nova tabela em um arquivo CSV
chi_despesas.to_csv('despesas_china_interp.csv', index=False)
```

```python
# Selecionar apenas as colunas desejadas
chi_despesas1 = chi[['Nome_do_pais', 'Ano', 'Despesa_em_pesquisa_e_desenvolvimento', 'Gasto_governamental_saude', 'Despesa_governamental_em_educacao']]

# Selecionar apenas os anos maiores ou iguais a 1995
chi_despesas1 = chi_despesas1[chi_despesas1['Ano'] >= 1995]

# Fazer a interpolação linear nos valores faltantes
chi_despesas_interpolated1 = chi_despesas1.interpolate()

# Converter colunas numércias para float
cols_numeric = ['Despesa_em_pesquisa_e_desenvolvimento', 'Gasto_governamental_saude', 'Despesa_governamental_em_educacao']
chi_despesas_interpolated1[cols_numeric] = chi_despesas_interpolated1[cols_numeric].astype(float)

# Exibir a nova tabela
chi_despesas_interpolated1.head(31)
```

```python
# Criar dicionário com novos nomes das colunas
new_cols = {'Despesa_em_pesquisa_e_desenvolvimento': 'P&D', 
            'Gasto_governamental_saude': 'Saúde', 
            'Despesa_governamental_em_educacao': 'Educação'}

# Renomear as colunas
chi_despesas_interpolated1 = chi_despesas_interpolated1.rename(columns=new_cols)

# Selecionar colunas renomeadas para o gráfico
cols = list(new_cols.values())

# Plotar gráfico de linhas
chi_despesas_interpolated1.plot(x='Ano', y=cols, kind='line', figsize=(10,6))

# Definir título e legendas dos eixos
plt.title('Estimativa da Despesas da china em P&D, saúde e educação (1995-2022)', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Despesa (% do PIB)', fontsize=12)

# Colocar legenda fora do gráfico
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

# Exibir gráfico
plt.show()

```

```python
# Criar uma nova coluna com a década correspondente a cada ano
chi['Decada'] = ((chi['Ano'] // 10) * 10).astype(str) + 's'

# Agrupar os dados por década e calcular a média das outras colunas
chi_por_decada = chi.groupby('Decada').mean()

# Exibir a tabela resultante
chi_por_decada.head(10)

chi_por_decada.to_csv('chi_decada.csv', index=False)
```
